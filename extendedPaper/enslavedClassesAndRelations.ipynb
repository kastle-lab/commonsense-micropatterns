```python
import rdflib
import os
from rich import print as rp

# Load Patterns into Graph

graph = rdflib.Graph()
patterns_folder = '../enslaved-modl/patterns/'

for filename in os.listdir(patterns_folder):
    if filename.endswith('.ttl'):
        file_path = os.path.join(patterns_folder, filename)
        graph.parse(file_path, format='turtle')

rp(f'Loaded {len(graph)} triples from {patterns_folder}')
```

```python
from rdflib.namespace import RDF, RDFS

# Get all classes from the graph
classes = set(graph.subjects(RDF.type, RDFS.Class))

rp(f'Found {len(classes)} classes:')

with open('classes.out', 'w') as f:
    for c in classes:
        f.write(f'{c.split("#")[-1]}\n')
        print(f'  {c.split("#")[-1]}')

```

```python
# Get all relations (predicates) from the graph

predicates = []
for s,p,o in graph.triples((None, None, None)):
    if p not in predicates:
        predicates.append(p)

rp(f'Found {len(predicates)} predicates:')
with open('relations.out', 'w') as f:
    for p in predicates:
        f.write(f'{p.split("#")[-1]}\n')
        print(f'  {p.split("#")[-1]}')
```
