```python
import os


summary = []
csmodl_patterns_folder = "../csmodl/patterns/"
enslaved_patterns_folder = "./enslaved-modl/patterns/"
output_folder = "./output/"
```

```python
# CS-MODL: Process Pattern Files


# Initialize lists to store class and property terms
csmodl_ClassTerms = []
csmodl_uniqueClassTerms = []
csmodl_PropTerms = []
csmodl_uniquePropTerms = []


# Iterate over each file in the patterns folder
for filename in os.listdir(csmodl_patterns_folder):
     file_path = os.path.join(csmodl_patterns_folder, filename)
     
     if os.path.isfile(file_path):
          with open(file_path, "r") as file:
               for line in file:

                    # Check for class terms
                    if "a rdfs:Class" in line:
                         term = line.split(" ")[0].split(":")[1]
                         term = "".join(e for e in term if e.isalnum())
                         
                         if term not in csmodl_ClassTerms:
                              csmodl_ClassTerms.append(f"{term}\t{filename}\t{1}")
                              csmodl_uniqueClassTerms.append(term)
                         else:
                              for i in range(len(csmodl_ClassTerms)):
                                   if term in csmodl_ClassTerms[i]:
                                        count = int(csmodl_ClassTerms[i].split("\t")[2])
                                        count += 1
                                        csmodl_ClassTerms[i] = f"{term}\t{filename}\t{count}"
                                        break
                    
                    # Check for property terms
                    elif "a rdf:Property" in line:
                         property = line.split(" ")[0].split(":")[1]
                         property = "".join(e for e in property if e.isalnum())
                         
                         if property not in csmodl_PropTerms:
                              csmodl_PropTerms.append(f"{property}\t{filename}\t{1}")
                              csmodl_uniquePropTerms.append(property)
                         else:
                              for i in range(len(csmodl_PropTerms)):
                                   if property in csmodl_PropTerms[i]:
                                        count = int(csmodl_PropTerms[i].split("\t")[2])
                                        count += 1
                                        csmodl_PropTerms[i] = f"{property}\t{filename}\t{count}"
                                        break


#Write class terms to a file
with open(f"{output_folder}csmodl_ClassTerms.tsv", "w") as file:
     for term in csmodl_ClassTerms:
          file.write(f"{term}\n")
     file.close()


# Create a dictionary to count class terms
csmodl_ClassTerm_counts = {}
for term in csmodl_ClassTerms:
     term_name = term.split("\t")[0]
     filename = term.split("\t")[1].split(".")[0]

     if term_name in csmodl_ClassTerm_counts:
          csmodl_ClassTerm_counts[term_name][0] += 1
          csmodl_ClassTerm_counts[term_name][1].append(filename)
          csmodl_ClassTerm_counts[term_name][1].sort()
     else:
          csmodl_ClassTerm_counts[term_name] = [1, [filename]]



# Write class term counts to a file
with open(f"{output_folder}csmodl_classTermCounts.tsv", "w") as file:
     for term in csmodl_ClassTerm_counts:
          if csmodl_ClassTerm_counts[term][0] > 1:
               file.write(f"{term}\t{csmodl_ClassTerm_counts[term]}\n")
file.close()

# Write property terms to a file
with open(f"{output_folder}csmodl_PropTerms.tsv", "w") as file:
     for term in csmodl_PropTerms:
          file.write(f"{term}\n")
     file.close()

# Create a dictionary to count property terms
csmodl_PropTerm_counts = {}
for term in csmodl_PropTerms:
     term_name = term.split("\t")[0]
     filename = term.split("\t")[1].split(".")[0]
     if term_name in csmodl_PropTerm_counts:
          csmodl_PropTerm_counts[term_name][0] += 1
          csmodl_PropTerm_counts[term_name][1].append(filename)
          csmodl_PropTerm_counts[term_name][1].sort()
     else:
          csmodl_PropTerm_counts[term_name] = [1, [filename]]

# Write property term counts to a file
with open(f"{output_folder}csmodl_PropTermCounts.tsv", "w") as file:
     for term in csmodl_PropTerm_counts:
          if csmodl_PropTerm_counts[term][0] > 1:
               file.write(f"{term}\t{csmodl_PropTerm_counts[term]}\n")
file.close()

print(f"shared classTermCounts: {len(csmodl_ClassTerm_counts)}")
print(f"uniqueClassTerms: {len(csmodl_uniqueClassTerms)}")
print(f"shared propTermCounts: {len(csmodl_PropTerm_counts)}")
print(f"uniquePropTerms: {len(csmodl_uniquePropTerms)}")
```

```python
# ENSLAVED-MODL

# Initialize lists to store class and property terms
enslaved_ClassTerms = []
unique_enslaved_ClassTerms = []
enslaved_PropTerms = []
unique_enslaved_PropTerms = []

# Iterate over each file in the patterns folder
for filename in os.listdir(enslaved_patterns_folder):
    file_path = os.path.join(enslaved_patterns_folder, filename)
    
    if os.path.isfile(file_path):
        with open(file_path, "r") as file:
            for line in file:
                # Check for class terms
                if "a rdfs:Class" in line:
                    term = line.split(" ")[0].split(":")[1]
                    term = "".join(e for e in term if e.isalnum())
                    
                    if term not in enslaved_ClassTerms:
                        enslaved_ClassTerms.append(f"{term}\t{filename}\t{1}")
                        unique_enslaved_ClassTerms.append(term)
                    else:
                        for i in range(len(enslaved_ClassTerms)):
                            if term in enslaved_ClassTerms[i]:
                                count = int(enslaved_ClassTerms[i].split("\t")[2])
                                count += 1
                                enslaved_ClassTerms[i] = f"{term}\t{filename}\t{count}"
                                break
                
                # Check for property terms
                elif "a rdf:Property" in line:
                    property = line.split(" ")[0].split(":")[1]
                    property = "".join(e for e in property if e.isalnum())
                    
                    if property not in enslaved_PropTerms:
                        enslaved_PropTerms.append(f"{property}\t{filename}\t{1}")
                        unique_enslaved_PropTerms.append(property)
                    else:
                        for i in range(len(enslaved_PropTerms)):
                            if property in enslaved_PropTerms[i]:
                                count = int(enslaved_PropTerms[i].split("\t")[2])
                                count += 1
                                enslaved_PropTerms[i] = f"{property}\t{filename}\t{count}"
                                break


# Write class terms to a file
with open(f"{output_folder}enslaved_ClassTerms.tsv", "w") as file:
    for term in enslaved_ClassTerms:
        file.write(f"{term}\n")
    file.close()

# Create a dictionary to count class terms
enslaved_ClassTerm_counts = {}
for term in enslaved_ClassTerms:
    term_name = term.split("\t")[0]
    filename = term.split("\t")[1].split(".")[0]
    if term_name in enslaved_ClassTerm_counts:
        enslaved_ClassTerm_counts[term_name][0] += 1
        enslaved_ClassTerm_counts[term_name][1].append(filename)
        enslaved_ClassTerm_counts[term_name][1].sort()
    else:
        enslaved_ClassTerm_counts[term_name] = [1, [filename]]

# Write class term counts to a file
with open(f"{output_folder}enslaved_classTermCounts.tsv", "w") as file:
    for term in enslaved_ClassTerm_counts:
        if enslaved_ClassTerm_counts[term][0] > 1:
            file.write(f"{term}\t{enslaved_ClassTerm_counts[term]}\n")
file.close()

# Write property terms to a file
with open(f"{output_folder}enslaved_PropTerms.tsv", "w") as file:
    for term in enslaved_PropTerms:
        file.write(f"{term}\n")
    file.close()

# Create a dictionary to count property terms
enslaved_PropTerm_counts = {}
for term in enslaved_PropTerms:
    term_name = term.split("\t")[0]
    filename = term.split("\t")[1].split(".")[0]
    if term_name in enslaved_PropTerm_counts:
        enslaved_PropTerm_counts[term_name][0] += 1
        enslaved_PropTerm_counts[term_name][1].append(filename)
        enslaved_PropTerm_counts[term_name][1].sort()
    else:
        enslaved_PropTerm_counts[term_name] = [1, [filename]]

# Write property term counts to a file
with open(f"{output_folder}enslaved_PropTermCounts.tsv", "w") as file:
    for term in enslaved_PropTerm_counts:
        if enslaved_PropTerm_counts[term][0] > 1:
            file.write(f"{term}\t{enslaved_PropTerm_counts[term]}\n")
file.close()

# Print the property terms and their count
print(f"Shared Class Terms: {len(enslaved_ClassTerm_counts)}")
print(f"Unique Class Terms: {len(unique_enslaved_ClassTerms)}")
print(f"Shared Prop Terms: {len(enslaved_PropTerm_counts)}")
print(f"Unique Prop Terms: {len(unique_enslaved_PropTerms)}")
```

```python
# SHARED TERMS

# Initialize lists to store shared terms
shared_classTerms = []
shared_propTerms = []
unique_classTerms = []
unique_propTerms = []

# Find unique class terms
for term in csmodl_ClassTerm_counts:
    if term not in unique_classTerms:
        unique_classTerms.append(term)
for term in enslaved_ClassTerm_counts:
    if term not in unique_classTerms:
        unique_classTerms.append(term)

# Find unique property terms
for term in csmodl_PropTerm_counts:
    if term not in unique_propTerms:
        unique_propTerms.append(term)
for term in enslaved_PropTerm_counts:
    if term not in unique_propTerms:
        unique_propTerms.append(term)

# Find shared class terms and calculate total counts
for term in csmodl_ClassTerm_counts:    
    if term in enslaved_ClassTerm_counts:
        enslavedCount = enslaved_ClassTerm_counts[term][0]
        enslavedTerms = enslaved_ClassTerm_counts[term][1]
        csmodlCount = csmodl_ClassTerm_counts[term][0]
        csmodlTerms = csmodl_ClassTerm_counts[term][1]
        totalCount = csmodlCount + enslavedCount

        shared_classTerms.append(f"{term}\t{csmodlCount}\t{csmodlTerms}\t{enslavedCount}\t{enslavedTerms}\t{totalCount}")

# Find shared property terms and calculate total counts
for term in csmodl_PropTerm_counts:
    if term in enslaved_PropTerm_counts:
        enslavedCount = enslaved_PropTerm_counts[term][0]
        enslavedTerms = enslaved_PropTerm_counts[term][1]
        csmodlCount = csmodl_PropTerm_counts[term][0]
        csmodlTerms = csmodl_PropTerm_counts[term][1]
        totalCount = csmodlCount + enslavedCount
        shared_propTerms.append(f"{term}\t{csmodlCount}\t{csmodlTerms}\t{enslavedCount}\t{enslavedTerms}\t{totalCount}")


# Write shared class terms to a file
with open(f"{output_folder}shared_classTerms.tsv", "w") as file:
    file.write("Term\tCS-MODL Count\tCS-MODL Files\tEnslaved-MODL Count\tEnslaved-MODL Files\tTotal Count\n")
    for term in shared_classTerms:
        file.write(f"{term}\n")

# Write shared property terms to a file
with open(f"{output_folder}shared_propTerms.tsv", "w") as file:
    file.write("Term\tCS-MODL Count\tCS-MODL Files\tEnslaved-MODL Count\tEnslaved-MODL Files\tTotal Count\n")
    for term in shared_propTerms:
        file.write(f"{term}\n")

# Print the number of shared terms
print(f"num_shared_classTerms: {len(shared_classTerms)}")
print(f"num_shared_propTerms: {len(shared_propTerms)}")

print(f"shared classTermCounts: {len(csmodl_ClassTerm_counts)}")
print(f"uniqueClassTerms: {len(csmodl_uniqueClassTerms)}")
print(f"shared propTermCounts: {len(csmodl_PropTerm_counts)}")
print(f"uniquePropTerms: {len(csmodl_uniquePropTerms)}")

print(f"Shared Class Terms: {len(enslaved_ClassTerm_counts)}")
print(f"Unique Class Terms: {len(unique_enslaved_ClassTerms)}")
print(f"Shared Prop Terms: {len(enslaved_PropTerm_counts)}")
print(f"Unique Prop Terms: {len(unique_enslaved_PropTerms)}")

# Create Summary
summary.append(f"Totals:")

summary.append(f"== CS-MODL ==")
summary.append(f"     CS-MODL Shared Class Terms: {len(csmodl_ClassTerm_counts)}")
summary.append(f"          * these class terms exist in more than one cs-modl pattern\n")
summary.append(f"     CS-MODL Unique Class Terms: {len(csmodl_uniqueClassTerms)}")
summary.append(f"          * this is the total count of unique class terms in all cs-modl patterns\n")

summary.append(f"     CS-MODL Shared Property Terms: {len(csmodl_PropTerm_counts)}")
summary.append(f"          * these property terms exist in more than one cs-modl pattern\n")
summary.append(f"     CS-MODL Unique Property Terms: {len(csmodl_uniquePropTerms)}")
summary.append(f"          * this is the total count of unique property terms in all cs-modl patterns\n")

summary.append(f"== Enslaved-MODL ==")

summary.append(f"     ENSLAVED-MODL Shared Class Terms: {len(enslaved_ClassTerm_counts)}")
summary.append(f"          * these class terms exist in more than one enslaved pattern\n")
summary.append(f"     ENSLAVED-MODL Unique Class Terms: {len(unique_enslaved_ClassTerms)}")
summary.append(f"          * this is the total count of unique class terms in all enslaved patterns\n")

summary.append(f"     ENSLAVED-MODL Shared Property Terms: {len(enslaved_PropTerm_counts)}")
summary.append(f"          * these property terms exist in more than one enslaved pattern\n")
summary.append(f"     ENSLAVED-MODL Unique Property Terms: {len(unique_enslaved_PropTerms)}")
summary.append(f"          * this is the total count of unique property terms in all enslaved patterns\n")

summary.append(f"== Across MODLs ==")

summary.append(f"     Shared Class Terms: {len(shared_classTerms)}")
summary.append(f"          * these class terms exist in both cs-modl and enslaved-modl patterns\n")
summary.append(f"     Unique Class Terms: {len(unique_classTerms)}")
summary.append(f"          * this is the total count of unique class terms in all patterns including cs-modl and enslaved-modl\n")

summary.append(f"     Shared Property Terms: {len(shared_propTerms)}")
summary.append(f"          * these property terms exist in both cs-modl and enslaved-modl patterns\n")
summary.append(f"     Unique Property Terms: {len(unique_propTerms)}")
summary.append(f"          * this is the total count of unique property terms in all patterns including cs-modl and enslaved-modl")

with open("summary.txt", "w") as file:
    for line in summary:
        print(line)
        file.write(f"{line}\n")
    file.close()
```
